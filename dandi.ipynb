## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana tren jumlah order per bulan dari Januari 2017 hingga Agustus 2018?


```python
all_df.sample(5)
```

```python
mask = (all_df["order_purchase_timestamp"] >= "2017-01-01") & \
       (all_df["order_purchase_timestamp"] <= "2018-12-31")
df_filtered = all_df.loc[mask]
```

```python
monthly_orders = df_filtered.resample(
    rule="M",
    on="order_purchase_timestamp"
).agg({"order_id": "nunique"})

# Drop bulan terakhir kalau tidak lengkap
max_date = df_filtered["order_purchase_timestamp"].max()
if max_date.day < max_date.days_in_month:
    monthly_orders = monthly_orders.iloc[:-1]

monthly_orders.index = monthly_orders.index.strftime("%Y-%m")
monthly_orders = monthly_orders.reset_index()
monthly_orders.rename(columns={"order_id": "order_count"}, inplace=True)
```

```python
monthly_orders
```

```python
max_value = monthly_orders['order_count'].max()
max_row = monthly_orders.loc[
    monthly_orders['order_count'].idxmax()
]

print("Bulan Tertinggi:", max_row['order_purchase_timestamp'])
print("Jumlah Order:", max_value)
```

```python
plt.figure(figsize=(12,6))
plt.plot(monthly_orders["order_purchase_timestamp"], monthly_orders["order_count"], marker="o")
plt.xticks(rotation=45)
plt.title("Jumlah Order per Bulan (Jan 2017 - Agustus 2018)")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Order")
plt.grid(True)
plt.show()
```

```python
orders_2017 = all_df[all_df['order_purchase_timestamp'].dt.year == 2017]
orders_2018 = all_df[all_df['order_purchase_timestamp'].dt.year == 2018]

total_2017 = orders_2017['order_id'].nunique()
total_2018 = orders_2018['order_id'].nunique()

growth = ((total_2018 - total_2017) / total_2017) * 100

print("Total Order 2017:", total_2017)
print("Total Order 2018:", total_2018)
print(f"Growth 2018 vs 2017: {growth:.2f}%")
```

### Pertanyaan 2: Segmen pelanggan mana yang paling berkontribusi terhadap revenue 2018?

FREQUENCY

```python
frequency_df = (
    orders_2018.groupby("customer_id")["order_id"]
    .nunique()
    .reset_index()
    .rename(columns={"order_id": "frequency"})
)
frequency_df['frequency_score'] = pd.qcut(
    frequency_df['frequency'].rank(method='first'),
    5,
    labels=[1,2,3,4,5]
).astype(int)
frequency_df.head()
```

**Insight:**
- Tren jumlah order per bulan menunjukkan pertumbuhan yang signifikan dari  tahun 2017 ke 2018.  
- Total transaksi meningkat dari **44.577 order** pada 2017 menjadi **53.769 order** pada 2018, atau tumbuh sebesar **20,62% year-over-year**.
- Puncak transaksi terjadi pada **November 2017** dengan **7.360 order**, yang mengindikasikan adanya lonjakan permintaan pada periode akhir tahun 2017.
- Sepanjang tahun 2018, volume order cenderung stabil di kisaran **6.0007.100 order per bulan**, menunjukkan performa bisnis yang lebih konsisten dibandingkan 2017.

 MONETARY

```python
monetary_df = (
    orders_2018.groupby("customer_id")["price"]
    .sum()
    .reset_index()
    .rename(columns={"price": "monetary"})
)
monetary_df['monetary_score'] = pd.qcut(
    monetary_df['monetary'], 5, labels=[1,2,3,4,5]
).astype(int)
monetary_df.head()
```

Recency

```python
recent_date = orders_2018["order_purchase_timestamp"].max()

recency_df = (
    orders_2018.groupby("customer_id")["order_purchase_timestamp"]
    .max()
    .reset_index()
    .rename(columns={"order_purchase_timestamp": "max_order_timestamp"})
)

recency_df["recency"] = (recent_date - recency_df["max_order_timestamp"]).dt.days
recency_df['recency_score'] = pd.qcut(
    recency_df['recency'], 5, labels=[5,4,3,2,1]
).astype(int)

recency_df.head()
```

Gabungkan jadi RFM

```python
rfm_df = (
    recency_df.merge(frequency_df, on="customer_id")
    .merge(monetary_df, on="customer_id")
)
rfm_df = rfm_df[['customer_id','recency_score','frequency_score','monetary_score']]

rfm_df.head()
```

```python
rfm_df['frequency_monetary_score'] = (
    rfm_df['frequency_score'] + rfm_df['monetary_score']
)// 2
rfm_df = rfm_df[['customer_id','recency_score','frequency_monetary_score']]
rfm_df.head()
```

```python
segment_dict = {
#hibernating
(1,1):"hibernating",
(1,2):"hibernating",
(2,1):"hibernating",
(2,2):"hibernating",
#about to sleep
(3,1):"about to sleep",
(3,2):"about to sleep",
# promising
(4,1):"promising",
# new customer
(5,1):"new customer",
# potential customer
(4,2):"potential customer",
(4,3):"potential customer",
(5,2):"potential customer",
(5,3):"potential customer",
# need atention
(3,3):"need atention",
# at risk
(1,3):"at risk",
(1,4):"at risk",
(2,3):"at risk",
(2,4):"at risk",
# do not lose
(1,5):"do not lose",
(2,5):"do not lose",
# loyal customer
(3,4):"loyal",
(3,5):"loyal",
(4,4):"loyal",
(4,5):"loyal",
# champions
(5,4):"champions",
(5,5):"champions"}
```

```python
rfm_df['segment']=rfm_df.apply(lambda x:
                                         segment_dict[(x['recency_score'],x['frequency_monetary_score'])]
                                        , axis= 1
                                        )
rfm_df.head()
```

```python
customer_segment_df = rfm_df.groupby(by="segment", as_index=False).customer_id.nunique().sort_values(by ="customer_id",ascending=False)
customer_segment_df
```

```python
customer_segment_df['segment'] = pd.Categorical(customer_segment_df['segment'], [
    "hibernating", "about to sleep", "promising",
    "new customer", "potential customer",
    "need atention","at risk","do not lose",
    "loyal","champions"
])
```

```python
top2 = customer_segment_df.nlargest(2, "customer_id")["segment"].tolist()
palette = {seg: "#72BCD4" if seg in top2 else "#D3D3D3" for seg in customer_segment_df["segment"]}
order = customer_segment_df.sort_values(by="customer_id", ascending=False)["segment"]
```

```python
plt.figure(figsize=(10, 5))

sns.barplot(
    x="customer_id",
    y="segment",
    data=customer_segment_df,
    order=order,
    palette=palette
)
plt.title("Number of Customer for Each Segment in 2018", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='y', labelsize=12)
plt.show()
```
